```python
from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()
import re
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[Lemmatizer.lemmatize(word) for word in review ]
    review=' '.join(review)
    corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
##simple preprocess is used to lower case the tokens 
```

```python
words = []
for sent in corpus :
    sent_token = sent_tokenize(sent)
    for sent in sent_token :
        words.append(simple_preprocess(sent))
```

```python
import gensim.downloader as api
import nltk
print("Loading Google News Word2Vec model...")
model = api.load("word2vec-google-news-300")  # takes a minute
## for better accuracy
```

```python
X = []
y_new = []  # to match X

for i in range(len(words)):
    vec = avg_word2vec(words[i])
    
    # if it's a valid vector
    if vec is not None and vec.shape[0] == model.vector_size:  # e.g., 300
        X.append(vec)
        y_new.append(y[i])
    else:
        print(f"Skipped: {i}, words: {words[i]}")
```

```python
X = np.array(X)
y = np.array(y_new)

print("Fixed X:", X.shape)
print("Fixed y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.ensemble import  RandomForestClassifier
random = RandomForestClassifier()
print(type(random))
```

```python
random.fit(X_train,y_train)
```

```python
y_pred = random.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
